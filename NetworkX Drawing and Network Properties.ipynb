```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
 G = nx.path_graph(5)
 nx.draw(G, with_labels=True)
 plt.show()
```

The above graph is a path graph with 4 edges and 5 nodes. The graph is not directed since it does not have any arrows to show direction and is not cyclic as well for any nodes.

```python
 G = nx.cycle_graph(5)
 nx.draw(G, node_color='green', with_labels=True)
 plt.show()
```

The above graph is a cycle graph with 5 nodes and 5 edges. The graph is not directed since it does not have any arrows to show direction and is cyclic.

```python
 G = nx.complete_graph(5)
 nx.draw(G, node_color='#A0CBE2', edge_color='red', width=2,
 with_labels=False)
 plt.show()
```

The above graph is a complete graph has 5 nodes and has 10 edges. The graph is not directed since it does not have any arrows.

```python
 G = nx.star_graph(5)
 pos=nx.spring_layout(G)
 nx.draw(G, pos, with_labels=True)
 plt.show()
```

The above graph is a star graph with 6 nodes and 5 edges. It is non directed since its edges dont have any arrows.

```python
 G = nx.lollipop_graph(3,2)
 nx.draw(G, with_labels=True)
 plt.show()
```

```python
list(nx.connected_components(G))
```

```python
nx.clustering(G)
```

There is only one connected component in the above graph. It is {0, 1, 2, 3, 4}.

Here we can see that nodes 0 and 1 have the highest clustering coefficient of 1 since they only have 2 neighbors and both of them are connected with each other which means the coefficient is 1/1 = 1. Next we can see that node 2 has a clustering coefficient of 0.333 since this node has 3 neigbors and only two neighbors are connected with each other out of 3 possible connections which gives us 1/3 = 0.3333. The other two ndoes have a clustering coefficent of 0 since their neighbors are not connected with each other.

```python
print("radius: %d" % nx.radius(G))
print("diameter: %d" % nx.diameter(G))
print("eccentricity: %s" % nx.eccentricity(G))
print("center: %s" % nx.center(G))
print("periphery: %s" % nx.periphery(G))
print("density: %s" % nx.density(G))
```

```python
print("Betweenness")
b = nx.betweenness_centrality(G)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, b[v]))
```

```python
print("Degree centrality")
d = nx.degree_centrality(G)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, d[v]))
```

```python
print("Closeness centrality")
c = nx.closeness_centrality(G)
for v in G.nodes():
    print("%0.2d %5.3f" % (v, c[v]))
```

```python
print("Eigenvector centrality")
centrality = nx.eigenvector_centrality(G)
sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
```

Node 2 has the highest Betweenness because it is present between two distinct groups and acts as a broker. 
Node 2 has the highest Degree centrality because it has the most number of links to other nodes ie has 3 connections.
Node 2 has the highest closeness centrality because it has the shortest average path length with all the toher nodes. This makes sense since 2 is placed in the center of the graph with minimal distance from all other nodes.
Node 2 has the highest eigen vector centrality since in lay man terms has the most popular neighbors. Here it has the highest weighted sum of its neighbor's centrality and hence is the highest in the network.

```python
pathlengths = []
print("source vertex {target:length, }")
for v in G.nodes():
    spl = dict(nx.single_source_shortest_path_length(G, v))
    print('{} {} '.format(v, spl))
    for p in spl:
        pathlengths.append(spl[p])
print('')
print("average shortest path length %s" % (sum(pathlengths) /
len(pathlengths)))
```

```python
dist = {}
for p in pathlengths:
    if p in dist:
        dist[p] += 1
    else:
        dist[p] = 1
print('')
print("length #paths")
verts = dist.keys()
for d in sorted(verts):
    print('%s %d' % (d, dist[d]))
```
